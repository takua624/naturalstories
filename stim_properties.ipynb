```python
import pandas as pd
```

```python
# investigate total word count and total character count


# load the Excel file
xls = pd.ExcelFile("parsed-natural-stories.xlsx")

total_words = 0
total_chars = 0

# iterate over each sheet
for sheet_name in xls.sheet_names:
    if len(sheet_name.split("_"))<2:
        continue

    print("Processing sheet:", sheet_name)
    this_words = 0
    this_chars = 0
    df = pd.read_excel(xls, sheet_name=sheet_name)
    if "Sentence" in df.columns:
        for sentence in df["Sentence"].dropna():
            word_count = len(str(sentence).split())
            this_words += word_count
            total_words += word_count
            char_count = len(str(sentence))
            this_chars += char_count
            total_chars += char_count
    print("Number of words in this sheet:", this_words)
    print("Number of characters in this sheet:", this_chars)
    print()

print("======\nTotal number of words across all sentences:", total_words)
print("Total number of characters across all sentences:", total_chars)
```
